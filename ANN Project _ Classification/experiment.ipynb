You should always do StandardScaling after train_test_split, not before.

Why is this wrong?

Because the scaler uses the mean and standard deviation of the entire dataset, including the test data → this is called data leakage.

```python
##Standardscalar

scalar=StandardScaler()
```

```python
x_train=scalar.fit_transform(x_train)  # fit + transform only on training
x_test=scalar.transform(x_test)   # transform using training params
```

1. What is fit()?
fit() = Learn / Calculate / Train

It learns the required parameters from the data, but does NOT return transformed output.

Examples:

StandardScaler.fit() → learns mean and std

CountVectorizer.fit() → learns vocabulary

PCA.fit() → learns principal components

You use fit() when you want to learn something from the dataset.
🔥 2. What is transform()?
transform() = Apply what was learned

It uses the parameters learned during fit() to transform new data.

Examples:

StandardScaler → applies the learned mean & std to scale the data

CountVectorizer → converts new text to vectors using learned vocab

PCA → projects new data onto learned components

You use transform() on:

training data (after fit)

test data

future data in production

🔥 3. What is fit_transform()?
fit_transform() = fit + transform (only on same data)

It does two things in one step:

Learns parameters from the data (fit)

Immediately applies transformation (transform)

When to use what? (Golden Rule)

✅ On training data

fit_transform()

✅ On test data

transform()    # ONLY transform, NEVER fit

❌ Never use fit() on test data


Because the test set must be like new unseen data.

| Method              | Meaning              | Learns Params | Returns Transformed Data | Use Case                       |
| ------------------- | -------------------- | ------------- | ------------------------ | ------------------------------ |
| **fit()**           | Learn/Calculate      | ✔️            | ❌                        | Only on training data          |
| **transform()**     | Apply learned params | ❌             | ✔️                       | Training (after fit) + Testing |
| **fit_transform()** | fit + transform      | ✔️            | ✔️                       | Only for training data         |
