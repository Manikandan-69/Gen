```python
import re
import nltk
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import SnowballStemmer
```

```python
snow_stem=SnowballStemmer('english')
```

```python
for i in range(len(data)):
    print(data['message'][i])
```

```python
## Removed all the special words and letters in the messages by using regular expression module
for i in range(len(data)):
    spl_rem=re.sub('[^A-Za-z]',' ',data['message'][i])
    print(spl_rem)
    
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(data)):
    spl_rem=re.sub('[^A-Za-z]',' ',data['message'][i])
    low_sen=spl_rem.lower()
    sent2word=nltk.word_tokenize(low_sen)
    ##print(sent2word)
    word_stem=[lemma.lemmatize(word,pos='v') for word in sent2word if word not in stopwords.words('english')]
    final_word=' '.join(word_stem)
    corpus.append(final_word)
```

```python
corpus
```

```python
## create a Bag of Words using max_feature parameter
##Binary BOW - Output as 1 and 0
##Normal BOW - Count will be get updated based on frequency
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(binary=False,max_features=50) ##For this dataset i choose max_feature as 50, but in larger dataset it may vary eg.2500
```

```python
cv.fit_transform(corpus)
```

```python
X=cv.fit_transform(corpus).toarray()
```
