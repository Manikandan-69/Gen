```python
## Stemming
## getting root word by removing preffix and suffix of its word
```

```python
words=['ate','eating','gone','got','congratulations','fairly','controlables','worried','is','was','do','does']
```

```python
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
```

```python
for word in words:
    print(word +'--->'+ stemming.stem(word))
```

```python
## PorterStemmer has not been get the root word properly.
```

```python
##Regexp Stemmer
```

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stem = RegexpStemmer('ing$|s$|e$|able$', min=3)
```

```python
for word in words:
    print(word +'--->'+ reg_stem.stem(word))
```

```python
## This Regular expression also has some disadvantages while steeming.
```

```python
##Snowball Stemmer
```

```python
from nltk.stem import SnowballStemmer
```

```python
snow_stem=SnowballStemmer('english')
```

```python
for word in words:
    print(word +'---->' + snow_stem.stem(word))
```

```python
## This snowball stemmer better than other stemmer with specific words which end like 'ly' . 
## But this also not produce a correct root word. 
```
