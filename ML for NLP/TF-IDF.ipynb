```python
import pandas as pd
```

```python
para=pd.read_csv('D:/GenAI/ML for NLP/Spam_Ham/spam_ham_large_dataset.csv')
```

```python
para
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(para)):
    rem_spl=re.sub('[^a-zA-Z0-9]',' ',para['message'][i])
    low_sent=rem_spl.lower()
    sent2word=nltk.word_tokenize(low_sent)
    rem_stp=[lem.lemmatize(word,pos='v') for word in sent2word if word not in stopwords.words('english')]
    final=' '.join(rem_stp)
    corpus.append(final)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=100,ngram_range=(1,1))
```

```python
tfidf.fit_transform(corpus)
```

```python
X=tfidf.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x: "%.3g" % x))
```

| Parameter                   | Meaning                                                  | What It Does                                             | Example Effect                                            |
| --------------------------- | -------------------------------------------------------- | -------------------------------------------------------- | --------------------------------------------------------- |
| **edgeitems=30**            | Number of items to show at start & end of each dimension | Prevents NumPy from hiding elements with `...` too early | Shows 30 values instead of default 3 before truncation    |
| **linewidth=100000**        | Max characters per printed line                          | Prevents line wrapping of large arrays                   | Large matrices print in **one line** instead of multiple  |
| **formatter={'float': fn}** | Custom print formatting for floats                       | Controls how floating-point numbers are displayed        | `"%.3g"` → 3 significant digits (e.g., `123.456` → `123`) |
