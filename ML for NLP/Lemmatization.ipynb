```python
from nltk.stem import WordNetLemmatizer
```

```python
words=['ate','eating','gone','got','congratulations','fairly','controlables','worried','is','was','do','does']
```

```python
lemma=WordNetLemmatizer()
```

```python
for word in words:
    print(word + ' --->' + lemma.lemmatize(word))
```

```python
import nltk
nltk.download('wordnet')
```

```python
for word in words:
    print(word + ' --->' + lemma.lemmatize(word))
```

```python
for word in words:
    print(word + ' --->' + lemma.lemmatize(word,pos='v'))
```

```python
## most of the word in the list are verbs , so I have given pos='v'. It gives me a correct result.
## According to the words we have to choose the pos and get the root word
```
