```python
import pandas as pd
```

```python
data=pd.read_csv('Spam_Ham/spam_ham_large_dataset.csv')
```

```python
data
```

```python
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer
lemma=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(len(data)):
    rem_spl=re.sub('[^A-Za-z]',' ',data['message'][i])
    low_sent=rem_spl.lower()
    split_sent=low_sent.split()
    #print(split_sent)
    final=[lemma.lemmatize(word,pos='v') for word in split_sent if word not in stopwords.words('english')]
    final=" ".join(final)
    corpus.append(final)
```

```python
corpus
```

Unigram

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=50,binary=True,ngram_range=(1,1))
```

```python
cv.fit_transform(corpus)
```

| Part                    | Meaning                         |
| ----------------------- | ------------------------------- |
| **200x50**              | 200 messages × 50 features      |
| **sparse matrix**       | only non-zero values stored     |
| **int64**               | integer counts                  |
| **689 stored elements** | non-zero entries                |
| **CSR format**          | compressed sparse row structure |


```python
X=cv.fit_transform(corpus).toarray()
```
